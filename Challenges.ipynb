```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

# 3. Valoración media por tienda

```python
tienda.head()
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
promedios = []

for i, df in enumerate(tiendas, start=1):
    promedio = float(df.Calificación.mean())
    promedios.append(promedio)
    print(f"Tienda {i}: {promedio:.2f}")
```

# 4. Productos más vendidos y menos vendidos

```python
tienda.head(3)
```

```python
import matplotlib.pyplot as plt
```

```python
#Tienda 1

masVendidos = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=True)

ax = masVendidos.plot(kind='barh', figsize=(10, 8), color='pink')

# Nota: ya no se usa ['Producto'] porque es una Serie
for i, v in enumerate(masVendidos):
    ax.text(v + 2, i, str(v), va='center', fontsize=10)

plt.title("Productos más vendidos por categoría - Tienda 1")
plt.xlabel("Cantidad vendida")
plt.ylabel("Categoría del Producto")
plt.show()
```

```python
#Tienda 2

masVendidos = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)

masVendidos.plot(
    kind='pie',
    figsize=(8, 8),
    autopct='%1.1f%%',   # porcentaje con un decimal
    startangle=90,
    colors=['pink', 'lightblue', 'lightgreen', 'lavender', 'lightcoral', 'khaki', 'Orange', 'Blue']
)

plt.title("Distribución de productos vendidos por categoría - Tienda 2")
plt.ylabel('')  # Oculta el texto del eje Y
plt.show()
```

```python
#Tienda 3

masVendidos3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=True)

ax = masVendidos3.plot(kind='barh', figsize=(10, 8), color='pink')

# Nota: ya no se usa ['Producto'] porque es una Serie
for i, v in enumerate(masVendidos3):
    ax.text(v + 2, i, str(v), va='center', fontsize=10)

plt.title("Productos más vendidos por categoría - Tienda 1")
plt.xlabel("Cantidad vendida")
plt.ylabel("Categoría del Producto")
plt.show()
```

```python
# Tienda 4


masVendidos = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)

ax = masVendidos.plot(kind='area', alpha=0.5, color='orange', figsize=(10, 6))

# 🔹 Agregar etiquetas con los valores en cada punto
for i, v in enumerate(masVendidos):
    plt.text(i, v + 5, str(v), ha='center', fontsize=9)

plt.title("Ventas por categoría (área) - Tienda 4")
plt.xlabel("Categoría del Producto")
plt.ylabel("Cantidad vendida")

plt.show()
```

# 6. Informe final: Análisis de tiendas para el Sr. Juan
En este análisis se buscó ayudar al Sr. Juan a decidir qué tienda vender entre sus cuatro sucursales. Para eso se revisaron varios factores: los ingresos totales, las categorías más y menos vendidas, las calificaciones de los clientes, los productos con mayor y menor venta y el costo de envío promedio. Con base en los datos y los gráficos hechos en el Colab, se presenta aquí un resumen con los resultados y una recomendación final.


## Desarrollo Ingresos totales
Al revisar los ingresos totales, se vio que las tiendas 1 y 3 son las que más dinero generan, mientras que la tienda 4 tiene los ingresos más bajos. Esto ya muestra que la tienda 4 no tiene un buen rendimiento general.

## Categorías de productos
En las categorías, las tiendas 1 y 3 también sobresalen, ya que venden más en las categorías principales. La tienda 2 está más o menos, y la tienda 4 otra vez se queda atrás, sin destacar en ninguna categoría importante.

## Calificaciones de clientes
Las calificaciones muestran algo similar: las tiendas 1 y 3 tienen los mejores puntajes, la tienda 2 es promedio, y la tienda 4 tiene la calificación más baja. Esto indica que los clientes no están tan satisfechos con la tienda 4.

## Productos más y menos vendidos
Los productos más vendidos están concentrados en las tiendas 1 y 3. En cambio, la tienda 4 tiene muchos productos de baja rotación y pocos que se vendan bien, lo cual afecta sus ingresos.

## Costo de envío promedio
La tienda 4 tiene un costo de envío un poco más bajo que las demás, pero este punto no compensa sus debilidades en ventas y calificaciones.

## **Conclusión**
Después de revisar todos los datos, se puede decir que la tienda 4 es la que se debería vender.
Aunque tiene costos de envío más bajos, sus ingresos son los menores, sus clientes están menos satisfechos y no destaca en ninguna categoría o producto.
Por el contrario, las tiendas 1 y 3 son las más fuertes, así que conviene mantenerlas.

En resumen, la mejor decisión para el Sr. Juan es vender la tienda 4 y conservar las demás, que son las que aportan más al negocio.
